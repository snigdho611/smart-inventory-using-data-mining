```python
#function that takes a CSV file with transactions and cleans it

def clean_data(file):
    groceries_DataFrame = pd.read_csv(file, nrows=51)
    groceries_DataFrame = groceries_DataFrame.fillna(0)
    return groceries_DataFrame
```

```python
#function that takes in a cleaned csv file and returns all the unique items in the dataset

def get_unique_data(cleaned_data):
    groceries_unique = list()
    length = len(cleaned_data.index)

    for i in range(0, len(cleaned_data.index)):
        for j in range(1, len(cleaned_data.iloc[i])):
            if cleaned_data.iloc[i][j] != 0 and cleaned_data.iloc[i][j] != '0':
                if cleaned_data.iloc[i][j] not in groceries_unique:
                    groceries_unique.append(cleaned_data.iloc[i][j])
                    
    groceries_unique_list = list()
    groceries_unique_list.append(['ID', 'Item', 'Price'])
    counter = 0
    for i in groceries_unique:
        groceries_unique_list.append([counter, i, str(round(random.random()*20, 2))])
        counter = counter + 1

    #print(groceries_unique_list)
    return groceries_unique_list
```

```python
#function to turn the cleaned dataset into one hot encoded data
#requires the list of all the unique items present in the original csv
#requires the original dataset to be cleaned

def one_hot_encode(unique_data, cleaned_data):
    groceries_DataFrameUnique = pd.DataFrame(unique_data)
    groceries_list = []
    for i in range(1, len(groceries_DataFrameUnique.index)):
        groceries_list.append(groceries_DataFrameUnique.iloc[i][1])
    groceries_list = [groceries_list]
    encoded_DataFrame = pd.DataFrame(index = cleaned_data.index, columns=groceries_list)
    encoded_DataFrame.fillna(value='0', inplace=True)
    
    for i in cleaned_data.index:
        for j in range(1, len(cleaned_data.iloc[i])):
            if cleaned_data.iloc[i][j] != 0 and cleaned_data.iloc[i][j] != '0':
                for k in list(encoded_DataFrame.columns):
                    if str(cleaned_data.iloc[i][j]) == str(k)[2:-3]:
                        encoded_DataFrame.loc[i, str(cleaned_data.iloc[i][j])]=1
    #print(groceries_list)
    
    return encoded_DataFrame
```
